In this notebook, you'll have an opportunity to play with $dt$ to see its effect on a simple model:

$\dot{y} = -y$

Given a small enough $dt$ value this model approximates $y = e^{-t}$ accurately.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = 12, 12
```

`simulate()` is the discrete version of the above differential equation.

```python
def simulate(y, dt):
    return y - y * dt
```

After the intial run, change the value of $dt$ to get a feel for its effect on the approximation. 

```python
total_time = 10
# DONE: change the value of dt
dt = 0.1
timesteps = np.arange(0, total_time, dt)
```

```python
# Initial value of y is 1
dt = 0.1
ys_01 = [1]

for _ in timesteps:
    y = simulate(ys_01[-1], dt)
    ys_01.append(y)
    
dt = 0.5
ys_05 = [1]

for _ in timesteps:
    y = simulate(ys_05[-1], dt)
    ys_05.append(y)

dt = 0.9
ys_09 = [1]

for _ in timesteps:
    y = simulate(ys_09[-1], dt)
    ys_09.append(y)

plt.plot(timesteps, ys_01[:-1], color='blue')
plt.plot(timesteps, ys_05[:-1], color='red')
plt.plot(timesteps, ys_09[:-1], color='green')

plt.ylabel('Y')
plt.xlabel('Time')
plt.title('Dynamics Model')
plt.legend()
plt.show()
```
